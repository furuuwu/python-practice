## math modes

LaTeX allows two writing modes for mathematical expressions: the inline math mode and display math mode:

* `inline math mode` is used to write formulas that are part of a paragraph

    the delimiters for inline mode are
    * `\(...\)` : (this one doesn't work here)
    * `$...$` : $...$
    * `\begin{math}...\end{math}` : (this one doesn't work here)

* `display math mode` is used to write expressions that are not part of a paragraph, and are therefore put on separate lines

    the delimiters for display mode are
    * `\[...\]` : (this one doesn't work here)
    * `\begin{displaymath}...\end{displaymath}` : (this one doesn't work here)
    * `\begin{equation}...\end{equation}` : \begin{equation}...\end{equation}

If you are typing a multi line mathematical process or derivation, it can be convenient to wrap the entire block in a `\begin` and `\end`. This allows us to forgo the `$` at the beginning and end of each line. Every line should end with a double foward slash `\\`.
