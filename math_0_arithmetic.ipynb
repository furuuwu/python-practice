# counting

The most ancient mathematical skill is counting, using, in the first instance, 
the natural numbers and later the integers
* the `natural numbers` is the set of the couting numbers $\mathbb{N} = \{1, 2, 3, \ldots\}$ or of the whole numbers $\mathbb{N}_0 = \{1, 2, 3, \ldots\}$, depending on who you ask
* the `integers` is the set $\mathbb{Z} = \{\ldots, -3, -2, -1, 0, 1, 2, 3, \ldots\} = \{0, 1, -1, 2, -2, \ldots\}$.

These sort of definitions are the "intuitive" ones. Formally these sets can be defined using equivalence classes (a topic for much later)

# the number line

The integers can be represented as equally spaced points on a line called the `number line`
* The set of all points (not just those representing integers) on the number line represents 
the `real numbers` $\mathbb{R}$. The general real number is usually denoted by the letter $x$ and we write $x \in \mathbb{R}$, meaning x belongs to the real numbers set (is a real number)
* A real number that can be written as the ratio of two integers, like 3/2 or âˆ’7/5 , is called a `rational number` $\mathbb{Q}$
* Other numbers, like $\sqrt{2}$ and $\pi$, that cannot be expressed in that way 
are called `irrational numbers` $\mathbb{R} \diagdown \mathbb{Q}$ (usually people don't give it a special symbol)

In computers, the integers and the rest of the reals have a different representation and are manipulated differently. Integers can be stored exactly but most reals can't. As will be seen later
* only numbers with finite binary representations (`dyadic/binary rational`) - numbers that can be expressed as a fraction whose denominator is a power of two - can be represented exactly
* because, you need to store every digit of the number and there is a finite amount of memory to store it
* eg. integers can always be expressed as a sum of powers of 2 so they are dyadic rationals. The fractions 1/2, 3/2, and 3/8 are dyadic rationals, but 1/3 is not
