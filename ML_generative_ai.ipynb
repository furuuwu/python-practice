#### text analysis

eg. sentiment classification - classify text (product reviews) as positive or negative

```python
# create a client object
client = openai.OpenAI()
# if the API key is not set as an env variable, you can pass it as an argument
# The api_key client option must be set either by passing api_key to the client
# or by setting the OPENAI_API_KEY environment variable
client
```

```python
def create_prompt(text):
    """ Create the input prompt for the language model 
    
    Args:
        text: text to classify
    
    Returns:
        prompt for text classification
    """
    instructions = 'Is the underlying sentiment positive or negative?'
    
    # define the format we want for the possible answers
    formatting = '"Positive" or "Negative"'
    
    return f'Text to classify: {text}\n {instructions} \n Answer({formatting}):'
```

```python
input_text = 'This project is cool'
```

```python
prompt = create_prompt(input_text)
prompt
```

https://platform.openai.com/docs/guides/chat-completions/overview

```python
def call_llm(prompt):
    # define the openai model to use
    
    # defining a sequence of prior messages in the chat is required to work with chat models
    messages = [
        {'content': prompt, 'role': 'user'} # 'user' means it has been specified by the user. Other opts: 'assistant', 'system'
    ]
    
    # use a chat model to complete my input messages
    # define the openai model to use
    model = 'gpt-4o-mini'
    response = client.chat.completions.create(messages=messages, model=model)
    
    return response
```

```python
answer = call_llm(prompt)
answer
```
