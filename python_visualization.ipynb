#### line plot

`plt.plot(t, y)` this creates a line plot (`matplotlib.lines.Line2D`)

create two sequences of x-values and y-values, and then use the general plot command
`plt.plot(x,y,[fmt],**kwargs)`
`[fmt]` is an optional string format and `**kwargs` are optional keyword arguments specifying the properties of the plot

```python
t = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(t)  # sine wave
```

```python
t[:10]
```

```python
y[:10]
```

```python
plt.plot(t, y)  # Line plot of t vs y
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.show()
```

as you can see, even tho the data is discrete points, it creates a line plot, giving the illusion of continuity

If you want other types of plots (eg., scatter plots), you need to specify options, like `"bo"` for blue circles without connecting lines, or use `plt.scatter()` for scatter plots

```python
plt.plot(t, y, 'x')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.show()
```

```python
plt.plot(t, y, marker='x')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sine Wave')
plt.show()
```

```python
"""
Maria throws a ball straight up with an initial speed v0 = 5 m/s.
Plot the position y(t) as a function of time
Numerically find (solve for) the maximum height reached by the ball
"""

"""
equations of motion
y = v0*t - (1/2)*g*t^2
v(t) = v0 - g*t
"""
```

```python
v0, g = 5, 9.8

t = np.linspace(0, 1.03, 100)  # create sequence of times
y = [v0 * u - g * u**2 / 2 for u in t]  # create sequence of positions, then plot y(t)
```

```python
plt.plot(t, y)
plt.show()
```

```python
plt.plot(t, y)
plt.title("Ball thrown straight up")  # add title to plot
plt.ylabel("Vertical distance y(t)")  # add labels for x and y axes
plt.xlabel("Time t [s]")
ymax = max(y)  # find position ymax
tmax = t[np.argmax(y)]  # find time tmax
print("Max height reached = ", f"{ymax:.4f}", " m")
print("Time to reach ymax = ", f"{tmax:.4f}", " s")
plt.show()
```

### matplotlib - 3d plots

```python
"""
plot the surface function z = f(x,y) = x+y+3 in 3D, 
when -3 ≤ x ≤ 3 and -3 ≤ y ≤ 3
"""
```

```python
x = np.arange(-3, 3, 0.6)  # grid of points on x-axis

X, Y = np.meshgrid(x, x)  # grid of points on xy-plane
# meshgrid(x, x) creates double arrays X,Y from the single NumPy array x.
# It is necessary to created double arrays which are required as inputs for the function plot_surface(X,Y,Z).

Z = 3 + X + Y  # values of z for points on the xy-grid
```

```python
# plot 3D surface using object oriented programming commands
# define the objects fig and axes for 3D plotting
fig = plt.figure()
axes = fig.add_subplot(projection="3d")

# plot 3D surface using wireframe style
axes.plot_wireframe(X, Y, Z, color="skyblue")

# customize the appearance of the plot
axes.set_xlabel("X")  # labels for x,y,z axes
axes.set_ylabel("Y")
axes.set_zlabel("Z")
axes.text(-2.6, 2, 6, "z=f(x,y)=3+x+y")  # add text to the plot
plt.show()
```

### matplotlib - 3d plots with `mpl_toolkits.mplot3d`

The mplot3d toolkit adds simple 3D plotting capabilities (scatter, surface, line, mesh, etc.) to Matplotlib by supplying an Axes object that can create a 2D projection of a 3D scene. The resulting graph will have the same look and feel as regular 2D plots. Not the fastest or most feature complete 3D library out there, but it ships with Matplotlib and thus may be a lighter weight solution for some use cases.

https://matplotlib.org/stable/api/toolkits/mplot3d.html

- scatter plot

`ax.scatter(x, y, z)`

- wireframe plot

`ax.plot_wireframe(x, y, z)`

- contour plot

`ax.contour3D(x, y, z, 50)`

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create a figure
fig = plt.figure()

# Add 3D subplot
ax = fig.add_subplot(111, projection="3d")

# Generate data
# Spherical coordinates
phi = np.linspace(0, np.pi, 100)
theta = np.linspace(0, 2 * np.pi, 100)
phi, theta = np.meshgrid(phi, theta)

# Cartesian coordinates
x = np.sin(phi) * np.cos(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(phi)

# Plot the surface
ax.plot_surface(x, y, z, color="b", alpha=0.6)

# Add labels
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

# Set the title
ax.set_title("3D Plot Example")

# Show the plot
plt.show()
```
